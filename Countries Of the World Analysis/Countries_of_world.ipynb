<li><b>Climate</b>: CO2 emissions (metric tons per capita) </li>

<li><b>Service</b>:  They include <b>value added </b>in wholesale and retail trade (including hotels and restaurants), transport, and government, financial, professional, and personal services such as education, health care, and real estate services to the GDP of the Country</li>

To narrow down the list, we want to find the variables that are significantly correlated with birthrate. 

### Statistical Calculation

Below, we will calculate the z-scores, and then the p-values, to find the correlations that are most significant, using a threshold of p < 0.05.

```python
# create a series that lists correlations between birthrate and all variables
birth = world_data.corr().Birthrate.drop('Birthrate')
```

```python
# calculate z-scores of the correlation of birth with all other variables
z_scores = stats.zscore(birth)
```

```python
# using the z-scores found above to calculate p-values
p_value = sts.norm.sf(abs(z_scores))
p_value = pd.Series(p_value).reset_index()
```

```python
birth = birth.reset_index()
```

```python
# merging dataframe with p-value
pd.merge(birth, p_value, right_index=True, left_index=True).drop(columns=['index_y', 'Birthrate']).\
            rename(columns={"index_x": "Variables", 0: "P_Value"}).nsmallest(3, 'P_Value')
```

After calculating the z-scores and p-values, we found that the only infant mortality and agriculture are significantly correlated with birthrate.

## On GDP

In this part we perform analysis of factors that can be possible causations of high or low GDP using Machine Learning.

Firstly, let's find the average GDP value around the world from this dataset.

```python
world_data.GDP.mean()
```

And remove variables that logically can`t be causations:

```python
df_gdp = world_data.drop(['Country','Region', 'Infant_mortality', 'Phones'], axis=1)
```

GDP of 10000 is taken for intial arranging of GDP values in 2 categories: high (>10000) and low (<10000)

```python
df_gdp['GDP_high'] = df_gdp.GDP > 10000
```

### Decision Tree

```python
X = df_gdp.drop(['GDP_high','GDP'], axis=1)
Y = df_gdp['GDP_high']
dt = tree.DecisionTreeClassifier(max_depth=3)
dt.fit(X,Y)
```

```python
dt_feature_names = list(X.columns)
dt_target_names = [str(s) for s in Y.unique()]
tree.export_graphviz(dt, out_file='tree.dot', 
    feature_names=dt_feature_names, class_names=dt_target_names,\
    filled=True) 
graph = pydotplus.graph_from_dot_file('tree.dot')
Image(graph.create_png())
```

### Findings from the above decision tree:

- Agriculture level is the main predictor that decision tree used to build the tree. Specifically, 6% is a borderline. Countries that have more than 6% of GDP from agriculture field - have low GDP as well.
- Another predictor is migration level. If migration level is dropping less than -4.015 (which means that more and more people emigrate from their countries) it is related to low GDP.
- Other predictors did't show anything interesting, and the third level of the decision tree unfortunately is not helpful.



Let's plot a barchart showing the relationship of agriculture level and GDP (High or Low).

```python
sns.barplot(x='GDP_high',y='Agriculture', data=df_gdp)
```

Also, for migration:

```python
sns.barplot(x='GDP_high',y='Migration', data=df_gdp)
```

The plot clearly shows that countries with high GDP have positive migration, and it is completely opposite for countries with low GDP.

Let's also do the decision tree analysis considering top 25 countries by GDP.

```python
world_data.GDP.nlargest(25)
```

```python
df_gdp['GDP_high'] = df_gdp.GDP >= 22000
```

```python
X = df_gdp.drop(['GDP_high','GDP'], axis=1)
Y = df_gdp['GDP_high']
dt = tree.DecisionTreeClassifier(max_depth=3)
dt.fit(X,Y)
```

```python
dt_feature_names = list(X.columns)
dt_target_names = ['Low','High']
tree.export_graphviz(dt, out_file='tree.dot', 
    feature_names=dt_feature_names, class_names=dt_target_names,\
    filled=True) 
graph = pydotplus.graph_from_dot_file('tree.dot')
Image(graph.create_png())
```

### Findings:
- Agriculture level is again the main predictor, but this time it dropped to 0.023. 
- Birthrate became a predictor. Top 25 countries by GDP have birthrates less than 12.755 (for every 1000 people)
- Other predictors are not very important or dividing only very small number of countries.

```python
sns.barplot(x='GDP_high',y='Agriculture', data=df_gdp)
```

```python
sns.barplot(x='GDP_high',y='Birthrate', data=df_gdp)
```

### Conclusion

The purpose of this part was to use a machine learning model to try to understand why some countries have high GDP and others don't, and also to find factors/variables that can be reasons. We built two decision trees, one is based on mean GDP as borderline between high and low, and the second one is about top 25 countries by GDP. To sum up from the above decision tree classification:

1. Pupulation, area, density, coastline: they have nothing to do with GDP. 

2. Main predictor is agriculture level. Countries with high GDP have very small level of agriculture. Top 25 countries by GDP have agriculture level no more than 2.3%, while in some countries with low GDP it can reach 30%.

3. Migration and birthrate are also factors that play important role in classification. Birthrate in countries with high GDP is around two times less compared to other countries. 13 births for 1000 people is the borderline for the top 25. Migration is positive in countries with high GDP and negative in countries with low GDP.

### Managerial insights

GDP is really complicated topic and there can be many factors that may cause GDP to raise or fall. This part is strongly based just on the data we have and the results of decision tree classifier. From the above analysis we can propose the following managerial insights:

- Countries that have very high agriculture level should focus on other fields. It doesn't mean that they need to just stop doing agriculture, although they need to develop other fields and develop educational system to improve percentage of highly-educated people that can improve those fields.

- Migration should be positive. Countries with low GDP need to create such conditions for their citizens that will lower emigration and prevent smart people from moving out of those countries. Easy to say, hard to do, but this is a very important factor. 

- Birthrate. Very obvious difference between countries with high GDP and low GDP. It's really hard to change birthrate statistics in any country, so giving any recommendations just based just on this data analysis not really possible. There is no country in top 25 GDP with high birthrate, so this aspect should be taken into serious consideration by governments of countries that are planning to achieve high GDP.
